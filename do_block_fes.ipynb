```python
import math
import sys
```

## Specify filename and all relevant constants: mins(R), maxs(R), number of bins, block_size

```python
filename = 'sR.weight'
smin = 1
smax = 112
nbin = 20
kbt = 2.494339
block_size = 10
```

```python
bin_size = (smax-smin)/float(nbin-1)
bin_size
```

## Read the file and do data binning of s(R); bin_index refers to the bin that a datapoint belongs to and bin_list contains bin indices of all datapoints

```python
bin_list=[]
w_list=[]
for lines in open(filename, "r").readlines():
    text = lines.strip().split()
    cv=float(text[0])
    bin_index = int( round( ( cv - smin ) / bin_size ) )
    w = float(text[1])
    bin_list.append(bin_index)
    w_list.append(w)
bin_list
```

```python
nblocks = int(len(bin_list)/block_size)
nblocks
```

```python
histoave = {} 
histoave2 = {}

for iblock in range(0, nblocks):
    start = iblock * block_size 
    end = start + block_size
    histo = {}
    for i in range(start, end):
        if bin_list[i] in histo: # If bin is already filled then sum up the bias
            histo[bin_list[i]] += w_list[i]
        else: # Add bias value to the bin if blin is empty                  
            histo[bin_list[i]]  = w_list[i] 
    # calculate average bias in each bin
    for key in histo: 
        histo[key] /= float(block_size)
    for key in histo:
        if key in histoave:
           histoave[key]   += histo[key]
           histoave2[key]  += histo[key] * histo[key]
        else:
           histoave[key]   = histo[key]
           histoave2[key]  = histo[key] * histo[key]

histoave
```

```python
log = open("fes."+str(block_size)+".dat", "w")
for i in range(0, nbin):
    # get CV values for that grid point
    s = smin + (float(i) * bin_size)
    log.write("%12.6lf " % s)
    # calculate fes
    nb = float(nblocks)  
    if i in histoave:
       # average and variance 
       aveh = histoave[i] / nb 
       s2h  = (histoave2[i]/nb-aveh*aveh) * nb / ( nb - 1.0 )
       # error
       errh = math.sqrt( s2h / nb )
       # free energy and error
       fes = -kbt * math.log(aveh)
       errf = kbt / aveh * errh 
       # printout
       log.write("   %12.6lf %12.6lf\n" % (fes, errf))
    else:
       log.write("       Infinity\n")
log.close()
```
